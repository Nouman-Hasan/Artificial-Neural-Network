```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
sns.set_theme(color_codes=True)
sns.set_style('whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv("kc_house_data.csv")
```

```python
df_wo_outliers = df.sort_values('price',ascending=False).iloc[216:]
```

```python
df_wo_outliers['date'] = pd.to_datetime(df_wo_outliers['date'])
```

```python
df_wo_outliers['month'] = df_wo_outliers['date'].apply(lambda date: date.month)
```

```python
df_wo_outliers['year'] = df_wo_outliers['date'].apply(lambda date: date.year)
```

```python
df_wo_outliers.drop('date', axis = 1, inplace=True)
```

```python
df_wo_outliers.drop('zipcode', axis = 1, inplace=True)
```

```python
df_wo_outliers['renovated'] = df_wo_outliers['yr_renovated'] > 0
```

```python
df_wo_outliers.drop(['yr_renovated'] ,axis = 1, inplace =True)
```

```python
df_wo_outliers.drop(['id'] ,axis = 1, inplace =True)
```

```python
df_wo_outliers['basement'] = df_wo_outliers['sqft_basement'] > 0
```

```python
df_wo_outliers['basement'] = pd.get_dummies(df_wo_outliers['basement'], drop_first=True)
```

```python
df_wo_outliers['renovated'] = pd.get_dummies(df_wo_outliers['renovated'], drop_first = True)
```

```python
df_wo_outliers.head()
```

```python
df_wo_outliers.drop('sqft_basement', axis = 1, inplace = True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_wo_outliers.drop('price' , axis = 1).values
y = df_wo_outliers['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler1 = MinMaxScaler()
```

```python
X_train = scaler1.fit_transform(X_train)
```

```python
X_test = scaler1.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam', loss = 'mse')
```

```python
model.fit(x=X_train, y=y_train,
          validation_data=(X_test,y_test),
          batch_size=128, epochs=600)
```

```python
df_wo_outliers['price'].describe()
```

```python
preds = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
np.sqrt(mean_squared_error(y_test,preds))
```
