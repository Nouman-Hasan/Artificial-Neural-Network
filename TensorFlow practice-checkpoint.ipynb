```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
sns.set_theme(color_codes=True)
sns.set_style('whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv('fake_reg.csv')
```

```python
df.head()
```

```python
sns.pairplot(data=df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('price',axis=1).values
```

```python
y = df['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(4,activation='relu')) #layer1
model.add(Dense(4,activation='relu')) #layer2
model.add(Dense(4,activation='relu')) #layer3

model.add(Dense(1)) # Output layer

model.compile(optimizer='rmsprop',loss='mse')
```

```python
model.fit(x = X_train, y = y_train, epochs=250)
```

```python
df_loss = pd.DataFrame(model.history.history)
```

```python
df_loss
```

```python
df_loss.plot()
```

```python
model.evaluate(X_train,y_train,verbose=0)
```

```python
model.evaluate(X_test,y_test,verbose=0)
```

```python
test_predictions = model.predict(X_test)
```

```python
test_predictions = pd.Series(test_predictions.reshape(300,))
```

```python
pred_df = pd.DataFrame(y_test,columns=['Test true Y'])
```

```python
pred_df = pd.concat([pred_df,test_predictions],axis = 1)
```

```python
pred_df.columns = ['Test true Y','Model Predictions']
```

```python
pred_df
```

```python
px.scatter(pred_df, x = 'Test true Y', y = 'Model Predictions', opacity=0.65,trendline='ols')
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
np.sqrt(mean_squared_error(pred_df['Test true Y'],pred_df['Model Predictions']))
```
